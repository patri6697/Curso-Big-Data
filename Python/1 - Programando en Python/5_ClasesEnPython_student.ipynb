# Ejemplo de clase Estudiante

Vamos a diseñar un programa que gestiona la lista de estudiantes de una asignatura y sus calificaciones. De cada estudiante guardaremos su nombre,
su grupo de teoría (que será la letra A, B o C), la nota obtenida en el examen y si ha entregado o no las prácticas. Tener aprobada la asignatura implica haber entregado las prácticas y haber obtenido en el examen una nota igual o superior a 5.
Deseamos hacer un programa que permita añadir estudiantes a la lista, mostrar la calificación de cada uno de ellos y efectuar algunas estadísticas sobre las notas, como obtener la nota media o el porcentaje de estudiantes que ha entregado las prácticas.

Definamos primero el tipo de datos Estudiante. Éste tiene 4 campos, el nombre, el grupo, la nota y la práctica.

```python
class Estudiante:
  def __init__(self, nombre, grupo, nota, practica):
    self.nombre = nombre
    self.grupo = grupo
    self.nota = nota
    self.practica = practica
```

Los campos nombre y grupo serán cadenas. Nota será un flotante y practica un booleano con *True* si entregó la práctica y *False* si no.

Sería interesante definir una función que leyera por teclado los datos de un estudiante y nos devolviera un nuevo objeto Estudiante con sus campos cumplimentados.

```python
def lee_estudiante():
  nombre = input('Nombre: ')
  grupo = input('Grupo (A/B/C): ')
  nota = float(input('Nota de examen: '))
  entregada = input('Práctica entregada (s/n): ')
  practica = entregada == 's'
  return Estudiante(nombre, grupo, nota, practica)
```

*lee_estudiante* es una función y no un método de la clase Estudiante.

```python
nuevo_estudiante = lee_estudiante()
```

El contenido de *nuevo_estudiante* será un objeto de la clase *Estudiante*.

Definamos ahora el método __str__ para poder imprimir en pantalla un estudiante:

```python
class Estudiante:
  def __init__(self, nombre, grupo, nota, practica):
    self.nombre = nombre
    self.grupo = grupo
    self.nota = nota
    self.practica = practica

  def __str__(self):
    cadena = 'Nombre: {0}\n'.format(self.nombre)
    cadena = cadena + 'Grupo: {0}\n'.format(self.grupo)
    cadena = cadena + 'Nota examen: {0:3.1f}\n'.format(self.nota)
    if self.practica:
      cadena = cadena + 'Práctica entregada'
    else:
      cadena = cadena + 'Práctica no entregada'
    return cadena
```

Ahora nos gustaría conocer la calificación de un estudiante: Matrícula de Honor, Notable, Aprobado o Suspenso. No existe un campo calificación en los objetos de la clase Estudiante, así que deberemos implementar un método que efectúe los cálculos pertinentes a partir del valor de práctica y del valor de nota:

```python
class Estudiante:
  def __init__(self, nombre, grupo, nota, practica):
    self.nombre = nombre
    self.grupo = grupo
    self.nota = nota
    self.practica = practica

  def __str__(self):
    cadena = 'Nombre: {0}\n'.format(self.nombre)
    cadena = cadena + 'Grupo: {0}\n'.format(self.grupo)
    cadena = cadena + 'Nota examen: {0:3.1f}\n'.format(self.nota)
    if self.practica:
      cadena = cadena + 'Práctica entregada'
    else:
      cadena = cadena + 'Práctica no entregada'
    return cadena

  def calificacion(self):
    if not (self.practica):
      return 'Suspenso'
    else:
      if self.nota < 5:
        return 'Suspenso'
      elif self.nota < 7:
        return 'Aprobado'
      elif self.nota < 8.5:
        return 'Notable'
      elif self.nota < 10:
        return 'Sobresaliente'
      else:
        return 'Matricula de Honor'
```

Probemos si funciona

```python
pepe = Estudiante('Pepe García', 'A', 7.7, True)
print(pepe.calificacion())
```

Si queremos obtener algunas estadísticas, como la nota media o el porcentaje de estudiantes que ha entregado las prácticas, definiremos y usaremos nuevas funciones:

```python
def nota_media(lista):
    suma = 0
    cantidad = 0
    for estudiante in lista:
      if estudiante.practica:
        suma += estudiante.nota
        cantidad += 1
      if cantidad != 0:
        return suma/cantidad
      else:
        return 0.0

def porcentaje_de_practicas_entregadas(lista):
  if len(lista) != 0:
    cantidad = 0
    for estudiante in lista:
      if estudiante.practica:
        cantidad += 1
    return cantidad / len(lista) * 100
  else:
    return 0.0
```
